```python
df = pd.DataFrame([271, 949, 1192, 1403, 5213], columns=["confirmed"], 
                 index=["Itaewon Clubs", "overseas inflow", "etc", "contact with patient", "Shincheonji Church"])
df
```

```python
f, ax = plt.subplots(1, 2, figsize=(18, 8)) # subplots는 2개의 리턴값이 있기 때문에 f라는 변수를 만듬.
# subplots(1, 2) 1개의 행에 2개의 그래프를 만든다.(인덱스 적용)
pie_chart.plot.pie(explode=[0.01, 0.01, 0.01, 0.01, 0.01], autopct="%1.2f%%",ax=ax[0])
# survived라는 컬럼의 값(1또는 0)을 가지고 가지고 pie차트를 만듬
ax[0].set_title('infection_place') # 제목과
ax[0].set_ylabel('') # 라벨

colors = ["purple", "orange", "blue", "green", "red"]
df.plot.bar(ax=ax[1], y='confirmed', color=colors) # countplot 은 막대형 그래프를 나타냄.
# raw_data를 사용할꺼고 그중에서 survived 컬럼을 사용할꺼다.
# subplots의 2번째 그래프는 sns(seaborn을 사용해서 인덱스 1번 순서에 넣어줌.)
ax[1].set_title('confirmed/place')
plt.show()
```
